# **Task 2: Data Cleaning & Feature Engineering**

# Formally check for missing values using .isnull().sum()

## Handle missing values appropriately:
#○ Fill missing country with "Unknown".
# ○ Fill missing director with "No Director Listed"

```python
df["country"] = df["country"].fillna("Unknown")
df["director"] = df["director"].fillna("No Director Listed")
```

```python
df
```

```python
df.isnull().sum()
```

# Extract a new numeric column from the duration field:
○ For Movies → extract minutes (e.g., "90 min" → 90)
○ For TV Shows → extract number of seasons (e.g., "2 Seasons" → 2)
○ Create:
■ duration_minutes (for movies)
■ seasons (for TV shows)
(use NaN for the irrelevant category)


```python
# First we create 2 empty comlumns names "duration_minutes" & "seasoms"

df["duration_minutes"] = np.nan
df["seasons"] = np.nan

for i, row in df.iterrows():
    if row["type"] == "Movie":
        if isinstance(row["duration"], str):
            df.at[i, "duration_minutes"] = int(row["duration"].split()[0])
    else:
        if isinstance(row["duration"], str):
            df.at[i, "seasons"] = int(row["duration"].split()[0])
```

```python
df
```

#Create a binary feature:
#○ Is_Recent = 1 if release_year >= 2015, otherwise 0.

```python
# we make a new column for it to store binary 0 and 1
df["Is_Recent"] = (df["release_year"] >= 2015).astype(int)
```

#Bivariate Analysis:
#Bar plot: Number of releases per country (top 10 only)
#○ Boxplot: Compare duration_minutes between movies from recent vs older years(is_Recent)

```python
#We make new column to count releases per country of top 10 so we use head(10) in value_counts() as it will give us top 10 countries with most releases
country_counts = df["country"].value_counts().head(10)
plt.figure(figsize=(14, 5))
sns.barplot(x=country_counts.index, y=country_counts.values)
plt.title("Number of Releases per Country (Top 10)")
plt.xlabel("Country")
plt.ylabel("Number of Releases")
plt.show()
```

# Correlation Heatmap:
# ○ Only for numerical features (e.g., release_year, duration_minutes, seasons,Is_Recent)

```python
corr=df[["release_year","duration_minutes","seasons","Is_Recent"]].corr()
```

```python
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

#Logic - this is a correlation heatmap of the 3 columns mentioned above basically light colour shows
# like correlation varies from -1 to +1 , -1 = negative corelation and +1 = positive correlation , 0 means no correlation as we can notice from the graph that correlation of column with itself will be +1 like we can see diagonally in the graph it shows 1
# Heatmap shows dark and light colors to show strentgh of correlation and in the right side number bar we can also see the colors correlation number(-1 to +1) too by the way the right side number is achieved by using annot=True
```

#Text Analysis Insight (Optional Bonus):
